##Q1: Revenue by Category

```python
import pandas as pd

url_q1 = 'https://raw.githubusercontent.com/MiguelNegron1/black_friday_sales_analysis/main/q1_revenue_per_category.csv'
df_q1 = pd.read_csv(url_q1)

from itertools import product
import matplotlib.pyplot as plt
import seaborn as sns


df_q1_sorted = df_q1.sort_values(by='category_revenue_usd', ascending=False)

plt.figure(figsize=(12,6))

sns.barplot(data=df_q1, x='product_category', y='category_revenue_usd', palette='viridis', hue='product_category', legend=False, order=df_q1_sorted.sort_values('category_revenue_usd', ascending=False)['product_category'] )


for index, row in df_q1_sorted.iterrows():
    label = f"${row['category_revenue_usd']/1e6:.1f}M" if row['category_revenue_usd'] >= 1e6 else f"${int(row['category_revenue_usd']):,}"
    plt.text(index, row['category_revenue_usd'] + 1e6, label, ha='center', fontsize=8)






plt.title('Revenue by Product Category (USD)', fontsize=14)
plt.xlabel('Product Categories')
plt.ylabel('Revenue (USD)')
plt.yscale('log')
plt.ylabel("Total Revenue ($, log scale)")
plt.tight_layout()

plt.show()
```

#  Insight #1 - Revenue by Product Category

* Category 1 generated the highest revenue, totaling $22.7 million USD.

* Category 8 followed with approximately $19.9 million USD.

* Category 5 came third, generating $16.4 million USD.

<br>

 **Recommendation:**
Focus marketing and inventory investment on the top-performing categories.
Consider reducing focus or phasing out low-performing ones like Category 20 and Category 19, which contribute negligibly to overall revenue.

##Q2: Do married customers purchase more products than unmarried ones, and does this trend differ by city tier?

```python
import pandas as pd

url_q2 = 'https://raw.githubusercontent.com/MiguelNegron1/black_friday_sales_analysis/main/q2_married_vs_unmarried.csv'
df_q2 = pd.read_csv(url_q2)

import matplotlib.pyplot as plt
import seaborn as sns



# Convert 0/1 to readable labels
df_q2['marital_status'] = df_q2['marital_status'].map({0: 'Unmarried', 1: 'Married'})

plt.figure(figsize=(10,6))
ax = sns.barplot(
    x='city_category',
    y='product_amount',
    hue='marital_status',
    data=df_q2,
    palette='Set2'
)

for container in ax.containers:
    ax.bar_label(container, fmt='{:,.0f}', label_type='edge', padding=3)

plt.title('Products Purchased by Marital Status and City Tier')
plt.xlabel('City Tier')
plt.ylabel('Total Products Purchased')
plt.legend(title='Marital Status')
plt.tight_layout()
plt.show()
```

#  Insight #2  Products Purchased by Marital Status and City Tier

Unmarried customers consistently purchase more products than married ones.

This trend holds across all city tiers, with the highest product volume seen in:

<br>

1.City Category B

2.City Category C

3.City Category A

<br>

 **Recommendation**:
Target advertisements toward unmarried individuals in City B, as this segment shows the strongest buying behavior. This could maximize return on marketing investment.



##Q3: How does age affect purchase behavior? Do different age groups spend differently?

```python
import pandas as pd

url_q3 = 'https://raw.githubusercontent.com/MiguelNegron1/black_friday_sales_analysis/main/q3_age_vs_behavior.csv'
df_q3 = pd.read_csv(url_q3)

import matplotlib.pyplot as plt
import seaborn as sns


df_q3_sorted = df_q3.sort_values(by='money_spent_usd', ascending=False)


plt.figure(figsize=(12,6))
ax = sns.barplot(
    x='age',
    y='money_spent_usd',
    palette='magma',
    data=df_q3,
    hue='age')

for index,row in df_q3.iterrows():
  label = f"${row['money_spent_usd'] / 1e6:.1f}M"
  ax.text(index, row['money_spent_usd'] +  1e6, label, ha='center', fontsize=9)

plt.title("Total Money Spent by Age Group (USD)", fontsize=14)
plt.xlabel("Age Group")
plt.ylabel("Total Spending (USD)")

plt.show()

```

#  Insight #3  Total Money Spent by Age Group


* The 2635 age group spent the most overall: $51.3 million USD.

* The 5155 age group had the highest average spending per product, at approximately $120.9 USD.

* Other strong age segments include 3645 and 1825, both with high total spending.

<br>

 **Recommendation**:

Tailor marketing campaigns toward both:



* The 2635 segment for volume-based promotions

* The 5155 segment for high-ticket items and premium upsells
